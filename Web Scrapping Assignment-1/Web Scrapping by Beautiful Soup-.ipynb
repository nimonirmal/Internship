# <span style="color:red">1) Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
page = requests.get('https://en.wikipedia.org/wiki/Main_Page') #to get the source code of the webpage 
soup = BeautifulSoup(page.content, 'html5lib')    #Parsing the HTML content

header_tags=[] #creating empty list
for i in soup.find_all("span",class_="mw-headline"): #extract the text from these tags one by looping over these tags
    header_tags.append(i.text) #adding text one by one to the list
    
#creating dataframe  
import pandas as pd
header=pd.DataFrame({"Headline":header_tags})
header
```

# <span style="color:red">2) Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
page = requests.get("https://presidentofindia.nic.in/former-presidents")
soup=BeautifulSoup(page.content,"html5lib")

name=[]
for i in soup.find_all("h3"):
    name.append(i.text)

    
term=[]
for i in soup.find_all("h5"):
    term.append(i.text)

import pandas as pd
former_presidents=pd.DataFrame({"Name of President":name,"Term of office":term})
former_presidents
```

# <span style="color:red">3) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame-


<span style="color:red">a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
page= requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup= BeautifulSoup(page.content)

Team=[]
Matches=[]
Points=[]
Rating=[]
Ranking=[]

for i in soup.find_all('tr')[:11]:
    cells = i.find_all('td')
    if len(cells) == 5:
        Ranking.append(cells[0].text.strip())
        Team.append(cells[1].text.strip().replace('\n'," , "))
        Matches.append(cells[2].text.strip())
        Points.append(cells[3].text.strip())
        Rating.append(cells[4].text.strip())
        
print('-'*50)
print("\t Top 10 ODI Teams in Men's Cricket")
print('-'*50)        


odi_team= pd.DataFrame({'Ranking':Ranking,'Team':Team,'Matches':Matches,'Points':Points,'Rating':Rating})  
odi_team
```

# <span style="color:red">5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
i) Headline            
ii) Time              
iii) News Link

```python
page = requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(page.content,"html5lib")

headline= []
for i in soup.find_all('a',class_="LatestNews-headline"):
    headline.append(i.text)    
    
time= []
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text) 
    
News_Link= []
for i in soup.find_all('a',class_="LatestNews-headline"):
    News_Link.append(i.get('href'))   
    
import pandas as pd

nf = pd.DataFrame({'headline':headline,'Time':time,'News Link':News_Link})
nf
```

# <span style="color:red">6) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
i) Paper Title                 
ii) Authors                  
iii) Published Date                       
iv) Paper URL

```python
page=requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
soup=BeautifulSoup(page.content,"html5lib")

title=[]
for i in soup.find_all("h2"):
    title.append(i.text)
author=[]
for i in soup.find_all("span",class_="sc-1w3fpd7-0 dnCnAO"):
    author.append(i.text)
date=[]
for i in soup.find_all("span",class_="sc-1thf9ly-2 dvggWt"):
    date.append(i.text)
url=[]
for i in soup.find_all("a",class_="sc-5smygv-0 fIXTHm"):
    url.append(i.get("href"))
    
import pandas as pd
pf = pd.DataFrame({'Titles':title,'Authors':author,'Published date':date,"URL":url})
pf

```

# <span style="color:red">7) Write a python program to scrape mentioned details from dineout.co.in and make data frame-
i) Restaurant name                   
ii) Cuisine                   
iii) Location                   
iv) Ratings                
v) Image URL

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup = BeautifulSoup(page.content)

titles = []
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text)    
cuisine = [] 
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])
location = [] 
for i in soup.find_all('div',class_="restnt-loc ellipsis"): 
    location.append(i.text)
rating = [] 
for i in soup.find_all('div',class_="restnt-rating rating-4"): 
    rating.append(i.text)
images = []
for i in soup.find_all("img",class_="no-img"):
    images.append(i.get('data-src'))
    
import pandas as pd

df = pd.DataFrame({'Titles':titles,'Cuisine':cuisine,'Location':location,'Rating':rating,'Images_url':images})
df
```
