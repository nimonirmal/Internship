# --------------------------------------------Q4------------------------------------

```python
from selenium import webdriver
driver4=webdriver.Chrome('chromedriver.exe')
driver4.get(" https://github.com/")
driver4.maximize_window()
time.sleep(5)
```

```python
#click on explore button
Explore = driver4.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button")
Explore.click()
```

```python
#Select trending option
trending = driver4.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a")
trending.click()
```

```python
#creating empty list
repo_urls = []  
rep_title = []
Description =[]
Contributors=[]
Language=[]
```

```python
#Scraping urls for each repository
repository = driver4.find_elements(By.XPATH,"//h1[@class = 'h3 lh-condensed']//a")
for i in repository[0:30]:
    repo_urls.append(i.get_attribute("href"))

    
#Scraping data from every repository page
for i in repo_urls:
    driver4.get(i)
    time.sleep(3)
    
    #Scraping description-
    try:
        title = driver4.find_element(By.XPATH,'//div[@class=" d-flex flex-wrap flex-items-center wb-break-word f3 text-normal"]')
        rep_title.append(title.text)
    except NoSuchElementException:
        rep_title.append('-')  
    
    #Scraping description-
    try:
        desc = driver4.find_element(By.XPATH,'//p[@class="f4 my-3"]')
        Description.append(desc.text)
    except NoSuchElementException:
        Description.append('-')
        
        
    #Scraping contributors count-
    try:
        cont_tags = driver4.find_element(By.XPATH,"//*[contains(text(),'    Contributors ')]")
        Contributors.append(cont_tags.text.replace('Contributors',''))
    except NoSuchElementException:
        Contributors.append('-')
        
        
    #Scraping Languages-
    try:
        cont_tags = driver4.find_element(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]')
        Language.append(cont_tags.text)
    except NoSuchElementException:
        Language.append('-')
```

```python
#print the length of column 
print(len(rep_title),len(Description),len(Contributors),len(Language))
```

```python
import pandas as pd
df = pd.DataFrame({'Repository title':rep_title,'Repository description':Description,
                   'Contributors count':Contributors,'Language used':Language})
df
```
