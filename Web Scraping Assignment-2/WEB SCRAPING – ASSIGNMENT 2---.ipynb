This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter ‚ÄúData Analyst‚Äù in ‚ÄúJob title, Skills‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note: All of the above steps have to be done in code. No step is to be done manually.

```python
# Initialize Chrome WebDriver
driver = webdriver.Chrome(r"C:\Users\nirma\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe")
# Open the shine.com website
driver.get("https://www.shine.com/")
#to maximize the window
driver.maximize_window()
```

```python
# entering designation and location as required in the questions-
designation = driver.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[1]/div/input")
designation.send_keys('Data Analyst')

location = driver.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input") 
location.send_keys('Bangalore')
```

```python
# create empty list for different attributes
Titles=[]
Location=[]
Company=[]
Experience=[]
```

```python
# Scraping Job-Titles's data
titles=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
for i in titles[:10]:
    Titles.append(i.text)
# Scraping Location's data
loc=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in loc[:10]:
    Location.append(i.text)    
# Scraping Company's data
com=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in com[:10]:
    Company.append(i.text)
# Scraping Experience's data
exp=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in exp[:10]:
    Experience.append(i.text)
```

```python
#creating dataframe for different attributes
import pandas as pd
Data_Analyst = pd.DataFrame({'Job Title':Titles,'Company':Company, 'Location':Location, 'Experience_Required':Experience})
Data_Analyst
```

```python
# Close the browser
driver.quit()
```

5. Now scrape data from this page as usual
6. Repeat this until you get data for 100sunglasses.
Note: That all of the above steps have to be done by coding only and not manually.

```python
# Initialize Chrome WebDriver
driver4 = webdriver.Chrome(r"C:\Users\nirma\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe")
# Open the flipkart.com website
driver4.get("https://www.flipkart.com/")
driver4.maximize_window()
```

```python
#clicking on pop_up to close
popupclose = driver4.find_element("xpath","/html/body/div[3]/div/span")
popupclose.click()
```

```python
#searching for sunglass 
search_product=driver4.find_element(By.CLASS_NAME,"Pke_EE")
search_product.send_keys("Sunglasses")
```

```python
#clicking on "search button"
search_button4= driver4.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/button")
search_button4.click()
```

```python
# create empty list for different attributes
brand=[]
product_desc=[]
price=[]
discount=[]
```

```python
start=0
end=3
for page in range(start,end):
    brand4=driver4.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brand4:
        brand.append(i.text)
    product=driver4.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in product:
        product_desc.append(i.text)
    product_price=driver4.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in product_price:
        price.append(i.text)                              
    price_off=driver4.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for i in price_off:
        discount.append(i.text)
    
    #to scrap data from next pages.
    next_button=driver4.find_element(By.XPATH,'//a[@class="_1LKTO3"]')  
    next_button.click()
    time.sleep(3)
```

```python
#creating dataframe for different attributes
import pandas as pd
sunglass=pd.DataFrame({})
sunglass['Brand name']=brand[:100]
sunglass['Product Description']=product_desc[:100]
sunglass['Price']=price[:100]
sunglass['Discount']=discount[:100]
```

```python
# Scraping Car's Names
name=driver10.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in name[:48]:
    car.append(i.text)

# Scraping Car's Price
Car_price=driver10.find_elements(By.XPATH,'//strong[contains(text(),"Price:")]')
for i in Car_price[:48]:
    price.append(i.text)
```
